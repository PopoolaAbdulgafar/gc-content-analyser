```python
from Bio import SeqIO
record = SeqIO.read("sequence.fasta", "fasta")
sequence = str(record.seq)
print(len(sequence))
import matplotlib.pyplot as plt

def slidingwindowgc(seq, window=1000):
    gc_values = []
    positions = []

    for i in range(0, 50000, window):
        fragment = seq[i:i+window]

        gc_count = fragment.count("G") + fragment.count("C")
        gc_percent = (gc_count / len(fragment)) * 100

        gc_values.append(gc_percent)
        positions.append(i)

    return positions, gc_values


positions, gc_values = slidingwindowgc(sequence)

plt.plot(positions, gc_values)
plt.xlabel("Position (bp)")
plt.ylabel("GC %")
plt.title("GC Content (First 50,000 bp)")
plt.savefig("gc_plot.png", dpi=300,bbox_inches='tight')
plt.show()
```
